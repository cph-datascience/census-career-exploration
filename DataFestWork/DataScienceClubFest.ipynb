```python
# Samuel Sparenga
# Data Science Club MiniFest
# April 26, 2026

import pandas as pd
import numpy as np

graduates_df = pd.read_csv ('../data/grad-students.csv')

graduates_df [0:10]
```

```python
ans_df = graduates_df [graduates_df ['Major_category'] == 'Agriculture & Natural Resources']
ans_df [0:10]
```

```python
bls_df = graduates_df [graduates_df ['Major_category'] == 'Biology & Life Science']
bls_df [0:10]
```

```python
cam_df = graduates_df [graduates_df ['Major_category'] == 'Computers & Mathematics']
cam_df [0:10]
```

```python
eng_df = graduates_df [graduates_df ['Major_category'] == 'Engineering']
eng_df [0:10]
```

```python
ics_df = graduates_df [graduates_df ['Major_category'] == 'Industrial Arts & Consumer Services']
ics_df [0:10]
```

```python
phy_df = graduates_df [graduates_df ['Major_category'] == 'Physical Sciences']
phy_df [0:10]
```

```python
# getting maximum salary for graph
# will utilize twice that number for graph 

import matplotlib.pyplot as plt
import seaborn as sns

graph_max = graduates_df ['Grad_median'].max () * 2
```

```python
labels = ["Agriculture & Natural Sciences",
          "Biology & Life Sciences",
          "Computing & Mathematics",
          "Engineering",
          "Industrial/Consumer Arts",
          "Physical Sciences"]

#box plots are vertically oriented
figure, (ax) = plt.subplots (6, sharex=True)

ax[0].set_ylabel (labels [0], rotation = 'horizontal', ha = 'right')
ax[0].boxplot (ans_df['Grad_median'], orientation = 'horizontal')

ax[1].set_ylabel (labels [1], rotation = 'horizontal', ha = 'right')
ax[1].boxplot (bls_df['Grad_median'], orientation = 'horizontal')

ax[2].set_ylabel (labels [2], rotation = 'horizontal', ha = 'right')
ax[2].boxplot (cam_df['Grad_median'], orientation = 'horizontal')

ax[3].set_ylabel (labels [3], rotation = 'horizontal', ha = 'right')
ax[3].boxplot (eng_df['Grad_median'], orientation = 'horizontal')

ax[4].set_ylabel (labels [4], rotation = 'horizontal', ha = 'right')
ax[4].boxplot (ics_df['Grad_median'], orientation = 'horizontal')

ax[5].set_ylabel (labels [5], rotation = 'horizontal', ha = 'right')
ax[5].boxplot (phy_df['Grad_median'], orientation = 'horizontal')

ax[0].set (title = "Box Plot of medians salaries by STEM category")
```
