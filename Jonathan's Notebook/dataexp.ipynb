## Which majors have the highest and lowest ShareWomen?
## What categories (e.g., Engineering vs Biology) tend to have more women?


```python
#Sort by ShareWomen, Group by Major_category and average ShareWomen
low_hi = women_stem.sort_values('ShareWomen').groupby('Major_category')['ShareWomen'].mean()
low_hi
```

```python
#### Bar chart: Top 10 (female-dominated STEM majors)
Top_10 = women_stem.sort_values('ShareWomen', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=Top_10, x='ShareWomen', y='Major_category', palette='magma')
plt.title('Top Female-Dominated Majors')
plt.show()
```

```python
#### Bar chart: Bottom 10 (male-dominated STEM majors)
Bottom_10 = women_stem.sort_values('ShareWomen', ascending=True).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=Bottom_10, x='ShareWomen', y='Major_category', palette='magma')
plt.title('Top Male-Dominated Majors')
plt.show()
```

```python
#### Boxplot: ShareWomen by Major_category
plt.figure(figsize=(12, 8))
sns.boxplot(data=women_stem, x='ShareWomen', y='Major_category')
plt.title('Distribution of Women Representation by Major Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# Graduation Advantage?
#### Do graduate degrees increase earnings?
#### Does this differ by major?
#### Are women concentrated in majors where graduating school matters more?
#### Top majors by Grad_premium
#### Female-dominated majors vs male-dominated majors in grad premium
#### Histogram of Grad_premium
#### Scatter: ShareWomen vs Grad_premium
#### Are women entering fields where advanced degrees are required to earn well?

#### Top majors by Grad_premium
#### Female-dominated majors vs male-dominated majors in grad premium
#### Histogram of Grad_premium
#### Scatter: ShareWomen vs Grad_premium
#### Are women entering fields where advanced degrees are required to earn well?

```python

# Use Major_code as key
df = grad.merge(
    r_grad[['Major_code', 'ShareWomen']],
    on='Major_code',
    how='left'
)

# Drop missing values (if any)
df = df.dropna(subset=['Grad_premium', 'ShareWomen'])



#  Do grad degrees increase earnings? 
mean_premium = df['Grad_premium'].mean()
print("Average Grad Premium:", round(mean_premium, 2))

# ---- Summary stats ----
print("\nGrad Premium Summary:")
print(df['Grad_premium'].describe())

#Top Majors By Grad Premium
top_majors = df.sort_values(by='Grad_premium', ascending=False)

print("\nTop 10 Majors by Grad Premium:")
print(top_majors[['Major', 'Grad_premium']].head(10))


##### Female vs Male Dominated Majors

female_dom = df[df['ShareWomen'] > 0.5]
male_dom   = df[df['ShareWomen'] < 0.5]

female_mean = female_dom['Grad_premium'].mean()
male_mean   = male_dom['Grad_premium'].mean()

print("\nAvg Grad Premium (Female-dominated majors):", round(female_mean, 2))
print("Avg Grad Premium (Male-dominated majors):", round(male_mean, 2))


#### Correlation

corr = df[['ShareWomen', 'Grad_premium']].corr().iloc[0,1]
print("\nCorrelation (ShareWomen vs Grad Premium):", round(corr, 3))


#### Histogram of Grad Premium

plt.figure(figsize=(8,5))
plt.hist(df['Grad_premium'], bins=30, edgecolor='black')
plt.title("Distribution of Grad Premium")
plt.xlabel("Grad Premium")
plt.ylabel("Frequency")
plt.axvline(mean_premium, color='red', linestyle='dashed', linewidth=1)
plt.show()



#### Compare by Major Category
category_premium = df.groupby('Major_category')['Grad_premium'].mean().sort_values()

plt.figure(figsize=(10,6))
category_premium.plot(kind='barh')
plt.title("Average Grad Premium by Major Category")
plt.xlabel("Average Grad Premium")
plt.ylabel("Major Category")
plt.show()


#### Highlight Female-heavy vs Premium

plt.figure(figsize=(8,5))

female = df[df['ShareWomen'] > 0.5]
male   = df[df['ShareWomen'] <= 0.5]

plt.scatter(female['ShareWomen'], female['Grad_premium'],
            color='purple', alpha=0.6, label='Female-dominated')

plt.scatter(male['ShareWomen'], male['Grad_premium'],
            color='blue', alpha=0.6, label='Male-dominated')

plt.title("Grad Premium vs Gender Composition")
plt.xlabel("Share Women")
plt.ylabel("Grad Premium")

plt.legend()   # <-- IMPORTANT (with parentheses)
plt.show()

#### Label top outliers
top5 = df.sort_values('Grad_premium', ascending=False).head(5)

plt.figure(figsize=(8,5))
plt.scatter(df['ShareWomen'], df['Grad_premium'], alpha=0.5)

for _, row in top5.iterrows():
    plt.text(row['ShareWomen'], row['Grad_premium'], row['Major'], fontsize=8)

plt.title("Top Grad Premium Majors (Labeled)")
plt.xlabel("Share Women")
plt.ylabel("Grad Premium")

plt.show()
```
