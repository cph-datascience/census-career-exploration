## Which majors have the highest and lowest ShareWomen?
## What categories (e.g., Engineering vs Biology) tend to have more women?


```python
#Sort by ShareWomen, Group by Major_category and average ShareWomen
low_hi = women_stem.sort_values('ShareWomen').groupby('Major_category')['ShareWomen'].mean()
low_hi
```

```python
#### Bar chart: Top 10 (female-dominated STEM majors)
Top_10 = women_stem.sort_values('ShareWomen', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=Top_10, x='ShareWomen', y='Major_category', palette='magma')
plt.title('Top Female-Dominated Majors')
plt.show()
```

```python
#### Bar chart: Bottom 10 (male-dominated STEM majors)
Bottom_10 = women_stem.sort_values('ShareWomen', ascending=True).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=Bottom_10, x='ShareWomen', y='Major_category', palette='magma')
plt.title('Top Male-Dominated Majors')
plt.show()
```

```python
#### Boxplot: ShareWomen by Major_category
plt.figure(figsize=(12, 8))
sns.boxplot(data=women_stem, x='ShareWomen', y='Major_category')
plt.title('Distribution of Women Representation by Major Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

## Representation to median salary
#### does gender dominance relate to pay?
#### Do majors with more women have lower median salaries?
#### Are male-dominated fields higher paying?


```python
#### Scatter plot: ShareWomen vs Median
sns.scatterplot(data=women_stem, x="ShareWomen", y="Median")
plt.show()
```

```python
#### Correlation: corr(ShareWomen, Median)
#### Is there a negative relationship between female representation and pay?
corr_matrix = women_stem.corr(numeric_only=True)
sns.heatmap(corr_matrix)
plt.show()
```

## Workforce outcomes
#### Do female-dominated majors have:
#### Higher unemployment?
#### More part-time work?
#### More low-wage jobs?

```python
"""
Unemployment_rate
Part_time / Full_time
Low_wage_jobs / Total
"""

r_grad['PartTimeRate'] = r_grad['Part_time'] / r_grad['Total']
r_grad['LowWageRate'] = r_grad['Low_wage_jobs'] / r_grad['Total']
```

```python
## Does Gender dominance influence job quality
#### Scatter: ShareWomen vs Unemployment_rate
plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    r_grad['ShareWomen'], 
    r_grad['Unemployment_rate'],
    c=r_grad['ShareWomen'],      
    cmap='coolwarm',             
    alpha=0.7
)

plt.colorbar(scatter, label='Share of Women')

plt.title('Gender Dominance vs Unemployment Rate')
plt.xlabel('Share of Women in Major')
plt.ylabel('Unemployment Rate')
plt.grid(True)
plt.show()
```

```python
#### Scatter: ShareWomen vs PartTimeRate

plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    r_grad['ShareWomen'], 
    r_grad['PartTimeRate'],
    c=r_grad['ShareWomen'],      
    cmap='coolwarm',             
    alpha=0.7
)

plt.colorbar(scatter, label='Share of Women')

plt.title('Gender Dominance vs Part Time Rate')
plt.xlabel('Share of Women in Major')
plt.ylabel('Part Time Rate')
plt.grid(True)
plt.show()
```

```python
####Scatter: ShareWomen vs LowWageRate

plt.figure(figsize=(10, 6))

scatter = plt.scatter(
    r_grad['ShareWomen'], 
    r_grad['LowWageRate'],
    c=r_grad['ShareWomen'],      
    cmap='coolwarm',             
    alpha=0.7
)

plt.colorbar(scatter, label='Share of Women')

plt.title('Gender Dominance vs Low Wage Rate')
plt.xlabel('Share of Women in Major')
plt.ylabel('Low Wage Rate')
plt.grid(True)
plt.show()
```
